```python
import pandas as pd #Biblioteca de Analise de Dados
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
base = pd.read_csv('trees.csv')
base.head()
```

```python
#Gráfico de dispersão considerando o volume e a dispesão
# método scatter gera um gráfico de dispersão
plt.scatter(x = base.Girth, y=base.Volume, color='Blue', facecolors = 'none', marker = '*')
plt.title('Arvores')
plt.xlabel('Circunferencia')
plt.ylabel('Volume')
```

```python
plt.plot(base.Girth, base.Volume)
plt.title('Arvores')
plt.xlabel('Circunferencia')
plt.ylabel('Volume')
```

```python
#Gráfico de dispersão com 'Afastamento' dos dados(jitter)
sns.regplot(x = base.Girth, y=base.Volume, data = base, x_jitter= 0.3, fit_reg= False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
base = pd.read_csv('co2.csv')
base.head()
```

```python
#Criando duas variaveis para da cada atributo (x = conc  e y = uptake)
x = base.conc
y = base.uptake
```

```python
# Retorna os valores únicos do atributo "treatment"
unicos = list(set(base.Treatment))
unicos
```

```python
for i in range(len(unicos)):
    indice = base.Treatment == unicos[i]
    plt.scatter(x[indice], y[indice], label = unicos[i])
plt.legend(loc = 'lower right')
```
