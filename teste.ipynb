```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

```python
df = pd.read_csv('leituraUm.csv', sep =';', dtype={'coluna1':int, 'coluna2': int, 'coluna3':int, 'coluna4':int})
```

```python
medias = df.mean()
```

```python
colors = ['green', 'orange','blue', 'gray']
plt.bar(medias.index, medias.values, color = colors)
plt.title('Média geral da prova - Primeiro trimestre')
plt.xlabel('Turmas')
plt.ylabel('Média')
plt.show()
```
